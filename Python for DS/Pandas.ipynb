### Advanced Operations Exercises


```python
# Exercise 1: Create a pivot table showing average profit by Product_Category and Customer_Segment
# First make sure we have the merged dataframe with all necessary columns
merged_df = pd.merge(market_df, customer_df, how='inner', on='Cust_id')
merged_df = pd.merge(merged_df, product_df, how='inner', on='Prod_id')

# Create pivot table
pivot_profit = pd.pivot_table(
    merged_df,
    values='Profit',
    index='Product_Category',
    columns='Customer_Segment',
    aggfunc='mean'
)

print("Pivot Table - Average Profit by Product Category and Customer Segment:")
print(pivot_profit)
```

```python
# Exercise 2: Find months with highest and lowest sales
# Extract month from Order_Date
orders_df['Month'] = pd.to_datetime(orders_df['Order_Date']).dt.month

# Merge with market facts to get sales data
sales_by_month = pd.merge(orders_df, market_df, on='Ord_id')
monthly_sales = sales_by_month.groupby('Month')['Sales'].sum().reset_index()

# Find highest and lowest sales months
highest_sales_month = monthly_sales.loc[monthly_sales['Sales'].idxmax()]
lowest_sales_month = monthly_sales.loc[monthly_sales['Sales'].idxmin()]

print("Monthly Sales Summary:")
print(monthly_sales)
print("\nMonth with Highest Sales:")
print(highest_sales_month)
print("\nMonth with Lowest Sales:")
print(lowest_sales_month)
```

```python
# Exercise 3: Calculate running total of profits ordered by date
# Sort by date
date_sorted_df = pd.merge(orders_df, market_df, on='Ord_id')
date_sorted_df['Order_Date'] = pd.to_datetime(date_sorted_df['Order_Date'])
date_sorted_df = date_sorted_df.sort_values('Order_Date')

# Calculate running total
date_sorted_df['Running_Profit'] = date_sorted_df['Profit'].cumsum()

print("Running total of profits by date:")
print(date_sorted_df[['Order_Date', 'Profit', 'Running_Profit']].head(10))

# Visualize the running profit
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(date_sorted_df['Order_Date'], date_sorted_df['Running_Profit'])
plt.title('Running Total of Profit Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Profit')
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()

```
