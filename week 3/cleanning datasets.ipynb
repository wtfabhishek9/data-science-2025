```python
import pandas as pd
import numpy as np

df = pd.read_csv("melbourne.csv")
df.head()
```

```python
# approx 23k rows, 21 columns
print(df.shape)
print(df.info())
```

```python
 # isnull() means a value is missing
 df.isnull()
```

```python
# summing up the missing values (column-wise)
df.isnull().sum()
```

```python
# columns having at least one missing value
df.isnull().any()

# above is equivalent to axis = 0 (by default, any() operates on columns)
df.isnull().any(axis = 1)
```

```python
 # rows having all missing values
 df.isnull().all(axis = 1)
```

```python
# sum it up to check how many rows have all missing values
df.isnull().all(axis = 1).sum()
```

```python
# sum of missing values in each row
df.isnull().sum(axis = 1)
```

```python
# summing up the missing values (column-wise)
round(100*(df.isnull().sum()/len(df.index)),2)  # percentage of missing values
```

```python
# removing the three columns with more than 50% missing values
df.dropna(thresh = len(df)*0.5, axis = 1, inplace = True)
round(100*(df.isnull().sum()/len(df.index)),2)  # percentage of missing values
```

```python
# removing the rows with more than 30% missing values
df.dropna(thresh = len(df.columns)*0.7, axis = 0, inplace = True)
round(100*(df.isnull().sum()/len(df.index)),2)  # percentage of missing values
```

```python
df[df.isnull().sum(axis = 1) > 3]  # rows with missing values
```

```python
# count the number of rows having > 3 missing values
# use len(df.index)
len(df[df.isnull().sum(axis = 1) > 3].index)
```

```python
# 177 rows having more than 3 missing values
# calculate the percentage of rows having more than 3 missing values
100 * len(df[df.isnull().sum(axis = 1) > 3] .index) / len(df.index)
```

```python
round(100*(df.isnull().sum()/len(df.index)),2)  # percentage of missing values
```

```python
# removing NaN Price rows
df = df[~np.isnan(df['Price'])]

round(100*(df.isnull().sum()/len(df.index)),2)  # percentage of missing values
```

```python
df['Landsize'].describe()
```

```python
# Removing NaN in Landsize
df = df[~np.isnan(df['Landsize'])]
round(100*(df.isnull().sum()/len(df.index)),2)  # percentage of missing values
```

```python
df.loc[:, ['Lattitude', 'Longtitude']].describe()
```

```python
# imputing Lattitude and Longtitude by mean values
df.loc[np.isnan(df['Lattitude']), ['Lattitude']] = df['Lattitude'].mean()
df.loc[np.isnan(df['Longtitude']), ['Longtitude']] = df['Longtitude'].mean()

round(100 * (df.isnull().sum() / len(df.index)),2)
```

```python
# remove CouncilArea Nan values
df = df[~df['CouncilArea'].isnull()]
round(100 * (df.isnull().sum() / len(df.index)),2)
```
